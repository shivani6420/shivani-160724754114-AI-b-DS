```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 25]

plt.figure(figsize=(6,4))
plt.bar(categories, y, color='orange', edgecolor='black')
plt.title("Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
