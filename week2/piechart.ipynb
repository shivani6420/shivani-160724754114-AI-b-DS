```python
import matplotlib.pyplot as plt

sizes = [25, 15, 30, 10, 20]
categories = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=categories, autopct='%1.1f%%',
        startangle=90,
        colors=['red','blue','green','orange','purple'])

plt.title("Pie Chart")
plt.show()



```
