```python
#Selecting Data
import pandas as pd
data={
    "ID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","David","Eva"],
    "Age":[22,25,23,24,26],
    "Marks":[85,78,92,88,90]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
print("Displaying only the 'Names' Column\n")
print(df["Name"])
print("\nDisplaying only the 'Name' and 'Marks' Columns\n")
print(df[["Name","Marks"]])
print("\nLocation 0\n",df.iloc[0])
print("\nLocation 1 to 4\n",df.iloc[1:4])
print("\nLocation 1 to 3\n",df.iloc[0:3,1:3])
```

```python
#Data Filtering
import pandas as pd
data={
    "ID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","David","Eva"],
    "Age":[22,25,23,24,26],
    "Marks":[85,78,92,88,90]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
print("\nFiltering data where Names are Alice or Charlie\n")
print(df[(df["Name"] == 'Alice') | (df["Name"] == 'Charlie')])
print("\nFiltering data where Marks is greater than 85\n")
print(df[df["Marks"]>85])
print("\nFiltering data where Age is less than 25\n")
print(df[df["Age"]<25])
```

```python
#Data Manipulation
import pandas as pd
data={
    "Name":["A","B","C"],
    "Age":[22,25,23],
    "Marks":[85,78,92]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
print("\nAdding a new column 'Bonus'\n")
df["Bonus"]=5
print(df)
print("\nModifying a column 'Marks'\n")
df["Marks"]=df["Marks"]+5
print(df)
print("\nCreating a new column based on other columns\n")
df["Total Marks"]=df["Marks"]+df["Bonus"]
print(df)
print("\nRenaming a column\n")
df.rename(columns={"Total":"Final"},inplace=True)
print(df)
print("\nDropping a column\n")
df.drop("Bonus",axis=1,inplace=True)
print(df)
print("\nChanging a particular value in the dataframe\n")
df.loc[0,"Age"]=21
print(df)
```

```python
#Selection Sort
n=int(input("Enter the number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}:")))
print(f"Original Array:{arr}")
for i in range(n-1):
  min_index=i
  for j in range(i+1,n):
    if arr[j]<arr[min_index]:
             min_index=j
  arr[i],arr[min_index]=arr[min_index],arr[i]
print("\nSorted Array:\n")
print(arr)
```

```python
#Bubble Sort
n=int(input("Enter the number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}:")))
print(f"Original Array:{arr}")
for i in range(n-1):
  for j in range(n-i-1):
    if arr[j ]>arr[j+1]:
      arr[j],arr[j+1]=arr[j+1],arr[j]
print("\nSorted Array:\n")
print(arr)
```

```python
# Insertion Sort
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

print("Original Array:", arr)

for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key

print("Sorted Array:", arr)
```

```python
#Linear Search
n=int(input("Enter the number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}:")))
key=int(input("\nEnter key element to be found:"))
found=False
for i in range(n):
  if arr[i]==key:
    print(f"Element found at position {i+1}")
    found=True
if found==False:
  print("Element not found")
```

```python
# Binary Search
n = int(input("Enter the number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))
arr.sort()
print("\nSorted list:", arr)
key = int(input("\nEnter key element to be found: "))
low = 0
high = n - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print(f"Element found at position {mid + 1}")
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
if not found:
    print("Element not found")
```
