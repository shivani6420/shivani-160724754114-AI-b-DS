```python

# Data Operations: Filtering, Manipulation, Ranking
import pandas as pd

# Load dataset from URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("Dataset (first 10 rows):")
print(df.head(10))

# -----------------------------
# a) Data Filtering
# Example: select only rows where species is 'versicolor'
filtered_data = df[df["species"] == "versicolor"]
print("\nFiltered Data (species = versicolor):")
print(filtered_data.head())

# -----------------------------
# b) Data Manipulation
# Example 1: Add a new column - Petal Ratio (petal_length / petal_width)
df["petal_ratio"] = df["petal_length"] / df["petal_width"]

# Example 2: Rename a column
df.rename(columns={"sepal_length": "sepal_len"}, inplace=True)

print("\nData after Manipulation (first 10 rows):")
print(df.head(10))

# -----------------------------
# c) Data Ranking
# Rank flowers based on sepal length (highest value = rank 1)
df["sepal_len_rank"] = df["sepal_len"].rank(ascending=False)

print("\nData with Ranking (sepal_len_rank) (first 10 rows):")
print(df[["sepal_len", "sepal_len_rank"]].head(10))



```
