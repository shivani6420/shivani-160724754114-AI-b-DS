```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = {'A': [10, 20, np.nan, 40],
        'B': [5, np.nan, 15, 20]}

df = pd.DataFrame(data)

# a) Show missing values
print("Missing Values:\n", df.isnull())

# b) Sum of missing values
print("\nSum of Missing Values:\n", df.isnull().sum())

# c) Replace missing values with column mean
df.fillna(df.mean(), inplace=True)

print("\nData after replacing with mean:\n", df)
```
