```python

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)  # X-axis: 1 to 5
y = np.array([10, 15, 7, 12, 9])  # Y-axis values
categories = ['A', 'B', 'C', 'D', 'E']

# -----------------------------
# a) Line Graph
plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o', color='blue', linestyle='-')
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
