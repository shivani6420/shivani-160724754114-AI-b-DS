```python
#Selecting Data
import pandas as pd
data={
    "ID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","David","Eva"],
    "Age":[22,25,23,24,26],
    "Marks":[85,78,92,88,90]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
print("Displaying only the 'Names' Column\n")
print(df["Name"])
print("\nDisplaying only the 'Name' and 'Marks' Columns\n")
print(df[["Name","Marks"]])
print("\nLocation 0\n",df.iloc[0])
print("\nLocation 1 to 4\n",df.iloc[1:4])
print("\nLocation 1 to 3\n",df.iloc[0:3,1:3])
```
